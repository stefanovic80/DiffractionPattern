\begin{eqnarray}
    I(\theta) & = & \left[ \frac{\sin\left( \frac{\pi a}{\lambda}\sin\theta \right)}{ \frac{\pi a}{\lambda}\sin\theta } \right]^2\qquad \sin\theta \approx \theta = \frac{x}{l}\qquad \Rightarrow\\
    I\left(\frac{x}{l}\right) & = & \left[ \frac{\sin\left( \frac{\pi a}{\lambda} \frac{x}{l} \right)}{ \frac{\pi a}{\lambda} \frac{x}{l} } \right]^2
\end{eqnarray}

```python
X = np.arange(-6, 6, 0.01)#*1e-2 #da -2 a +2 cm. Position on the screen
x = X*1e-2
l = 1#metres. Distance from slit to the screen
a = 5e-5 #slit size in metres (40 micrometres)
#480e-9, 'blue', 
wl = [400e-9, 'purple', 530e-9, 'green', 800e-9, 'red'] #wavelength size in metres and color

I_0 = 10^4 #Watt/m^2
```

```python
fig = plt.figure(figsize = (11, 4))
ax = fig.add_subplot(111)
plt.title('Diffraction')
ax.set_xlabel('x position on the screen (cm)', size = 15)
ax.set_ylabel('$I$ ($Watt/m^2$)', size = 15)
ax.set_xlim(-3, +3)
ax.grid(True)

I = []
k = 0
for j in range(0, len(wl), 2):
    I = I + [ I_0*( np.sin( np.pi*a*x/(wl[j]*l) )/( np.pi*a*x/(wl[j]*l) ) )**2 ]
    line, = ax.plot(X, I[k], Linewidth = 5, color = wl[ j + 1 ], label = str( int( wl[ j ]*1e9 ) ) + ' nm'  )
    k+=1
    
ax.legend()
"""I[0] = purple 400nm,I[1] = green 530nm I[2] = red 800nm"""
plt.show()
```

<b>Two slits diffraction Pattern</b>

```python
plotsNumber = int(len(wl)/2)

u = 800
shift = 130
c = len(X[:u])

for k in range(plotsNumber):
    fig = plt.figure(figsize = (11, 3))
    ax = fig.add_subplot(111)
    #plt.title('Diffraction Pattern')
    ax.set_xlabel('x position on the screen (cm)' )
    ax.set_ylabel('$I$ ($Watt/m^2$)' , size = 15)
    ax.set_xlim(-3, 3)
    ax.grid(True)

    Interference = I[k][:u] + I[k][shift:shift + u]

    line, = ax.plot(X, I[k], Linewidth = 5, color = wl[ 2*k + 1 ],\
                    label = str( int( wl[ j ]*1e9 ) ) + ' nm'  )
    line, = ax.plot(X[:u], I[k][shift:shift + u], Linewidth = 5, \
                    color = wl[ 2*k + 1 ], label = 'shifted slit' )
    line, = ax.plot(X[:u], Interference, Linewidth = 5, color = 'blue' )
    ax.legend()
    """I[0] = purple 400nm,I[1] = green 530nm I[2] = red 800nm"""
    plt.show()
```
