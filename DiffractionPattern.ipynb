\begin{eqnarray}
    I(\theta) & = & \left[ \frac{\sin\left( \frac{\pi a}{\lambda}\sin\theta \right)}{ \frac{\pi a}{\lambda}\sin\theta } \right]^2\qquad \theta & = & \frac{x}{l}\qquad \Rightarrow\\
    I(\theta) & = & \left[ \frac{\sin\left( \frac{\pi a}{\lambda} \frac{x}{l} \right)}{ \frac{\pi a}{\lambda} \frac{x}{l} } \right]^2
\end{eqnarray}

```python
X = np.arange(-3, 3, 0.01)#*1e-2 #da -2 a +2 cm. Position on the screen
x = X*1e-2
l = 1#metres. Distance from fenditure to the screen
a = 4e-5 #fenditure size in metres

wl = [400e-9, 'purple', 530e-9, 'green', 800e-9, 'red'] #wavelength size in metres and color

I_0 = 10^4 #Watt/m^2


fig = plt.figure(figsize = (11, 9))
ax = fig.add_subplot(111)
plt.title('Diffraction')
ax.set_xlabel('x position on the screen (cm)')
ax.set_ylabel('$I$ ($Watt/m^2$)')
ax.set_xlim(X[0], X[-1])
ax.grid(True)


j = 0

I = I_0*( np.sin( np.pi*a*x/(wl[j]*l) )/( np.pi*a*x/(wl[j]*l) ) )**2
line, = ax.plot(X, I, Linewidth = 5, color = wl[ j + 1 ] )

j = j + 2

I = I_0*( np.sin( np.pi*a*x/(wl[j]*l) )/( np.pi*a*x/(wl[j]*l) ) )**2
line, = ax.plot(X, I, Linewidth = 5, color = wl[ j + 1 ] )

j = j + 2

I = I_0*( np.sin( np.pi*a*x/(wl[j]*l) )/( np.pi*a*x/(wl[j]*l) ) )**2
line, = ax.plot(X, I, Linewidth = 5, color = wl[ j + 1 ] )

plt.show()
```
